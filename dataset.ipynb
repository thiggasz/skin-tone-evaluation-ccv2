# Imports and general definitions

```python
import pandas as pd
import matplotlib.pyplot as plt
from analysis.predictions_analysis import create_dataframe, plot_correlation_heatmap, plot_dispersion_tendency, plot_error_bias
from  analysis.predictions_analysis import plot_predictions_range, plot_quantitative_distribution, plot_error_boxplot
from analysis.results_analysis import get_matrix_file
```

```python
TRUE_FILE = ''
ANALYSIS_FILE = ' '
PRED_FILE = ''

scales_dict = {
    'fitzpatrick': ['type i', 'type ii', 'type iii', 'type iv', 'type v', 'type vi'],
    'monk': ['scale 01', 'scale 02', 'scale 03', 'scale 04', 'scale 05', 'scale 06', 'scale 07', 'scale 08', 'scale 09', 'scale 10']
}

df_matrix = get_matrix_file(PRED_FILE, 'fitzpatrick_type', False)
df_analysis = pd.read_csv(ANALYSIS_FILE, dtype={'subject_id': str})

df_complete = create_dataframe(df_matrix, df_analysis)
```

# Analysis functions

```python
def get_heatmaps(scale):    
    columns = ['error', 'luminance mean', 'luminance std', 'noise', 'sharpness', 'yaw', 'pitch', 'roll', 'contrast', 'spill', 'temperature']
    
    classes_list = scales_dict[scale]
    num_classes = len(classes_list)
    nrows = (num_classes + 2) // 2
    
    fig, axes = plt.subplots(nrows=nrows, ncols=2, figsize=(18, 5 * nrows))
    fig.suptitle(f'Heatmaps by {scale.capitalize()} Type', fontsize=18, fontweight='bold')

    ax_list = axes.flatten()
    
    for i, current_class in enumerate(classes_list):
        df = df_complete[df_complete['true_label'] == current_class]

        plot_correlation_heatmap(df, columns, ax=ax_list[i])

    for j in range(i + 1, len(ax_list)):
        fig.delaxes(ax_list[j])

    plt.tight_layout()
    plt.show()
```

```python
def get_error_bias(scale):    
    classes_list = scales_dict[scale]
    num_classes = len(classes_list)
    nrows = (num_classes + 2) // 2
    
    fig, axes = plt.subplots(nrows=nrows, ncols=2, figsize=(18, 5 * nrows))
    fig.suptitle(f'Error Bias by {scale.capitalize()} Type', fontsize=18, fontweight='bold')

    ax_list = axes.flatten()
    
    for i, current_class in enumerate(classes_list):
        df = df_complete[df_complete['true_label'] == current_class]

        plot_error_bias(df, ax=ax_list[i])

    for j in range(i + 1, len(ax_list)):
        fig.delaxes(ax_list[j])

    plt.tight_layout()
    plt.show()
```

```python
def get_dispersion_tendency(scale, column):    
    classes_list = scales_dict[scale]
    num_classes = len(classes_list)
    nrows = (num_classes + 2) // 2
    
    fig, axes = plt.subplots(nrows=nrows, ncols=2, figsize=(18, 5 * nrows))
    fig.suptitle(f'Error Bias by {scale.capitalize()} Type', fontsize=18, fontweight='bold')

    ax_list = axes.flatten()
    
    for i, current_class in enumerate(classes_list):
        df = df_complete[df_complete['true_label'] == current_class]

        plot_dispersion_tendency(df, attribute_column=column, ax=ax_list[i])

    for j in range(i + 1, len(ax_list)):
        fig.delaxes(ax_list[j])

    plt.tight_layout()
    plt.show()
```

```python
def get_error_boxplot(scale, column):    
    classes_list = scales_dict[scale]
    num_classes = len(classes_list)
    nrows = (num_classes + 2) // 2
    
    fig, axes = plt.subplots(nrows=nrows, ncols=2, figsize=(18, 5 * nrows))
    fig.suptitle(f'Error Bias by {scale.capitalize()} Type', fontsize=18, fontweight='bold')

    ax_list = axes.flatten()
    
    for i, current_class in enumerate(classes_list):
        df = df_complete[df_complete['true_label'] == current_class]

        plot_error_boxplot(df, attribute_column=column, ax=ax_list[i])

    for j in range(i + 1, len(ax_list)):
        fig.delaxes(ax_list[j])

    plt.tight_layout()
    plt.show()
```

```python
def get_predicions_range(scale, column):    
    classes_list = scales_dict[scale]
    num_classes = len(classes_list)
    nrows = (num_classes + 2) // 2
    
    fig, axes = plt.subplots(nrows=nrows, ncols=2, figsize=(18, 5 * nrows))
    fig.suptitle(f'Error Bias by {scale.capitalize()} Type', fontsize=18, fontweight='bold')

    ax_list = axes.flatten()
    
    for i, current_class in enumerate(classes_list):
        df = df_complete[df_complete['true_label'] == current_class]

        plot_predictions_range(df, attribute_column=column, ax=ax_list[i])

    for j in range(i + 1, len(ax_list)):
        fig.delaxes(ax_list[j])

    plt.tight_layout()
    plt.show()
```

```python
def get_distribution(scale, column):    
    classes_list = scales_dict[scale]
    num_classes = len(classes_list)
    nrows = (num_classes + 2) // 2
    
    fig, axes = plt.subplots(nrows=nrows, ncols=2, figsize=(18, 5 * nrows))
    fig.suptitle(f'Error Bias by {scale.capitalize()} Type', fontsize=18, fontweight='bold')

    ax_list = axes.flatten()
    
    for i, current_class in enumerate(classes_list):
        df = df_complete[df_complete['true_label'] == current_class]

        plot_quantitative_distribution(df, attribute_column=column, ax=ax_list[i])

    for j in range(i + 1, len(ax_list)):
        fig.delaxes(ax_list[j])

    plt.tight_layout()
    plt.show()
```
